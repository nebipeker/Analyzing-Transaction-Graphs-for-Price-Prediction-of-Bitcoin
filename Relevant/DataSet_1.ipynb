```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel('Merged.xlsx').set_index('Date')
```

```python
data
```

```python
data["Number Of Transactions"]
```

```python
plt.figure(figsize = (15, 10))
plt.plot(data["Number Of Transactions"])
plt.xlabel('dates')
plt.title('Number Of Transactions')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.plot(data["Bitcoin Estimated Transaction Volume"])
plt.xlabel('dates')
plt.title('Bitcoin Estimated Transaction Volume')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.plot(data["Bitcoin Cost Per Transaction"])
plt.xlabel('dates')
plt.title('Bitcoin Cost Per Transaction')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.plot(data["Price"])
plt.xlabel('dates')
plt.title('Price')
plt.show()
```
