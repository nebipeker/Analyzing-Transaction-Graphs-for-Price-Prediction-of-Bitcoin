### ANALYSIS

* General Connection

```python
nx.degree(us_graph)
```

```python
type(nx.degree(us_graph))
```

* Degree of Connection

```python
nx.degree(us_graph,"McCain")
```

* Finding most important Node

```python
nx.degree_centrality(us_graph)
```

```python
type(nx.degree_centrality(us_graph))
```

```python
nx.degree_centrality(us_graph).values()
```

```python
sorted(nx.degree_centrality(us_graph).values())
```

```python
most_importants = nx.degree_centrality(us_graph)
```

```python
for w in sorted(most_importants, key= most_importants.get, reverse=True):
    print(w,most_importants[w])
```

#### Connection Importancy

```python
nx.eigenvector_centrality(us_graph)
```

```python
most_important_links = nx.eigenvector_centrality(us_graph)
```

```python
for w in sorted(most_important_links, key= most_important_links.get, reverse=True):
    print(w,most_important_links[w])
```

### Betweenness Centrality
* In graph theory, betweenness centrality is a measure of centrality in a graph based on shortest paths.

```python
nx.betweenness_centrality(us_graph)
```

```python
best_connecter = nx.betweenness_centrality(us_graph)

for w in sorted(best_connecter, key= best_connecter.get, reverse=True):
    print(w,best_connecter[w])
```

### Building Subgroup
* Breadth-First Search
